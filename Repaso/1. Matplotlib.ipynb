# Ejemplo de gráficos con Matplotlib por Einer Zapata

```python
# importación de numpy
import numpy as np
# se importa pyplot de matplotlib
from matplotlib import pyplot as plt
```

```python
# Generamos los datos para el gráfico
x = np.linspace(1, 50, 500)

# se ejecuta la funcion seno para cada uno de los valores de x
y = np.sin(x)

# se ejecuta la funcion coseno para cada uno de los valores de x
y_2 = np.cos(x)
```

```python
# Creamos el gráfico

# para evitar el bloqueo de jupyter
plt.ion()
# se realiza el grafico
plt.plot(x,y,'b--')

# titulo
plt.title("Gráfico de Seno")

# se agregan los valores para las coordenadas
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
```

```python
# Creamos un grafico comparativo

# para evitar el bloqueo de jupyter
plt.ion()
# se realiza el grafico
plt.plot(x,y,'b',x,y_2,'r')

#La leyenda de cada grafico
plt.legend(['Seno','Coseno'])

# titulo
plt.title("Comparación de Seno y Coseno")

# se agregan los valores para las coordenadas
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
```

```python
# Grafico multiple

# Creamos el gráfico

# se define el tamaño y el numero de subgraficos
# en este caso dos gráficos en cada eje
plt.subplots(2,2,figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(x,y,'k--')
plt.legend(['Seno'])
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
plt.title("Representacion de Seno")

plt.subplot(2,2,4)
plt.plot(x,y_2,'y*')
plt.legend(['Coseno'])
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
plt.title("Representacion de Coseno")


plt.show()
```
